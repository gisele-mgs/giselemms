## **Descrição**
O **Covid-19 Dataset** contém dados globais sobre a pandemia de Covid-19, coletados por diversas fontes oficiais, como a Organização Mundial da Saúde (OMS), John Hopkins University e outros. Este dataset fornece informações sobre casos, mortes e recuperação por país e região, bem como métricas relacionadas à propagação da doença.

- **Fonte**: [Our World in Data (GitHub)](https://github.com/owid/covid-19-data)  
- **Número de Registros**: Variável, dependendo da data de atualização, mas normalmente contém mais de 200 registros por país/território.  
- **Número de Atributos**: Aproximadamente 50 colunas, dependendo da versão mais recente do dataset.  

## **Atributos**:
1. `iso_code` - Código do país (ex: `USA`, `BRA`).
2. `continent` - Continente do país (ex: Europa, Ásia, América do Norte).
3. `location` - Nome do país/território.
4. `date` - Data de coleta dos dados.
5. `total_cases` - Número total de casos confirmados de Covid-19.
6. `new_cases` - Número de novos casos confirmados no dia.
7. `total_deaths` - Número total de mortes devido à Covid-19.
8. `new_deaths` - Número de novas mortes no dia.
9. `total_recovered` - Número total de indivíduos recuperados.
10. `new_recovered` - Número de novos recuperados no dia.
11. `active_cases` - Número de casos ativos.
12. `critical_condition` - Número de casos em estado crítico.
13. `total_tests` - Número total de testes realizados.
14. `tests_per_million` - Número de testes realizados por milhão de pessoas.
15. `population` - População total do país.

# 1.Teste de Kolmogorov-Smirnov

```python
from scipy import stats
from scipy.stats import shapiro, kstest, normaltest, norm, ttest_ind

mean, std = (covid_data['total_deaths'].dropna().mean(), covid_data['total_deaths'].dropna().std())

stat_ks, p_ks = kstest(covid_data['total_deaths'].dropna(), 'norm', (mean, std))

print(f"Teste de Kolmogorov-Smirnov: Estatística = {stat_ks}, p-valor = {p_ks}")
if p_ks > 0.05:
    print("Os dados seguem uma distribuição normal (Não rejeita a Hipótese Nula).\n")
else:
    print("Os dados não seguem uma distribuição normal (Rejeita a Hipótese Nula).\n")

```

# 2.Teste de Shapiro-Wilk

```python

shapiro_stat, shapiro_pvalue = stats.shapiro(covid_data['total_deaths'].dropna())

print(f"Shapiro-Wilk: Estatística = {shapiro_stat:}, p-valor = {shapiro_pvalue}")

if shapiro_pvalue > 0.05:
    print("Os dados seguem uma distribuição normal (Não rejeita a Hipótese Nula).")
else:
    print("Os dados não seguem uma distribuição normal (Rejeita a Hipótese Nula).")
```

# 3. Teste de D'Agostino-Pearson



```python
stat_dp, p_dp = normaltest(covid_data['total_deaths'].dropna())
print("Teste de D'Agostino-Pearson:")
print(f"Estatística: {stat_dp:.4f}, p-valor: {p_dp:.4f}")
if p_dp > 0.05:
    print("Os dados seguem uma distribuição normal (Não rejeita a Hipótese Nula).\n")
else:
    print("Os dados não seguem uma distribuição normal (Rejeita a Hipótese Nula).\n")
```

# Teste a hipótese:



H0: Países do continente europeu não têm uma média significativamente maior do que países da América do Norte

H1: Países do continente europeu têm uma média de mortes por milhão significativamente maior do que países da América do Norte

```python
#Filtrar os continentes

alpha = 0.05
europa = covid_data[covid_data['continent'] == 'Europe'] ['total_deaths_per_million']
eua = covid_data[covid_data['continent'] == 'North America'] ['total_deaths_per_million']

#Verificar tipo de distribuição
teste_europa = stats.shapiro(europa)[1] > 0.05
teste_eua = stats.shapiro(eua)[1] > 0.05

#print(f"Teste de Shapiro-Wilk para Europa: {teste_europa}")
#print(f"Teste de Shapiro-Wilk para América do Norte: {teste_eua}")

if teste_europa and teste_eua:
    # Realizar o teste t-Student para amostras independentes
    t_stat, p_value = stats.ttest_ind(europa, eua)
    print(f"t-Student: Estatística = {t_stat}, p-valor = {p_value}")


    if p_value < alpha:
        print("Rejeitamos H₀: Países do continente Europeu têm uma média de mortes por milhão significativamente maior do que países da América do Norte.")
    else:
        print("Não rejeitamos H₀: Países do continente europeu não têm uma média significativamente maior do que países da América do Norte.")
else:
    # Realizar o teste Mann-Whitney se os dados não forem normais
    mannwhitney_stat, mannwhitney_pval = stats.mannwhitneyu(europa, eua)
    print(f"Mann-Whitney: Estatística = {mannwhitney_stat}, p-valor = {mannwhitney_pval}")

    if mannwhitney_pval < alpha:
        print("Rejeitamos H₀: Países do continente europeu têm uma média de mortes por milhão significativamente maior do que países da América do Norte.")
    else:
        print("Não rejeitamos H₀: Países do continente europeu não têm uma média significativamente maior do que países da América do Norte.")


```
