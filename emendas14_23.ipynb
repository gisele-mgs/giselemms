```python
df_filtrado["valor_empenhado"] = df_filtrado["Valor Empenhado"]
df_filtrado["localidade_gasto"] = df_filtrado["Localidade do gasto"]
```

```python
def obter_regiao(localidade):
    if localidade in ['ACRE (UF)', 'AMAPÁ (UF)', 'AMAZONAS (UF)', 'PARÁ (UF)',
                      'RONDÔNIA (UF)', 'RORAIMA (UF)', 'TOCANTINS (UF)']:
        return 'Norte'
    elif localidade in ['ALAGOAS (UF)', 'BAHIA (UF)', 'CEARÁ (UF)', 'MARANHÃO (UF)',
                        'PARAÍBA (UF)', 'PERNAMBUCO (UF)', 'PIAUÍ (UF)',
                        'RIO GRANDE DO NORTE (UF)', 'SERGIPE (UF)']:
        return 'Nordeste'
    elif localidade in ['DISTRITO FEDERAL (UF)', 'GOIÁS (UF)', 'MATO GROSSO (UF)',
                        'MATO GROSSO DO SUL (UF)']:
        return 'Centro-Oeste'
    elif localidade in ['ESPÍRITO SANTO (UF)', 'MINAS GERAIS (UF)', 'RIO DE JANEIRO (UF)',
                        'SÃO PAULO (UF)']:
        return 'Sudeste'
    elif localidade in ['PARANÁ (UF)', 'RIO GRANDE DO SUL (UF)', 'SANTA CATARINA (UF)']:
        return 'Sul'

df_filtrado['regiao'] = df_filtrado['localidade_gasto'].apply(obter_regiao)
```

# Teste de Aderência

```python
from scipy import stats
from scipy.stats import shapiro, kstest, normaltest, norm, ttest_ind

df['Valor Empenhado'] = pd.to_numeric(df['Valor Empenhado'], errors='coerce')

mean, std = df['Valor Empenhado'].mean(), df['Valor Empenhado'].std()  # Média e desvio padrão dos dados

# Now apply the kstest on the numeric column
stat_ks, p_ks = kstest(df['Valor Empenhado'].dropna(), 'norm', args=(mean, std))
print("Teste de Kolmogorov-Smirnov:")
print(f"Estatística: {stat_ks:.4f}, p-valor: {p_ks:.4f}")
if p_ks > 0.05:
    print("Os dados seguem uma distribuição normal (não rejeita H0).\n")
else:
    print("Os dados não seguem uma distribuição normal (rejeita H0).\n")
```

```python
stat_dp, p_dp = normaltest(df['Valor Empenhado'].dropna())
print("Teste de D'Agostino-Pearson:")
print(f"Estatística: {stat_dp:.4f}, p-valor: {p_dp:.4f}")
if p_dp > 0.05:
    print("Os dados seguem uma distribuição normal (Não rejeita a Hipótese Nula).\n")
else:
    print("Os dados não seguem uma distribuição normal (Rejeita a Hipótese Nula).\n")
```

# Análise de Variância

**Hipótese Alternativa (H1):** O valor total pago nas emendas é influenciado significativamente pela localidade do gasto.

**Hipótese Nula (H0):** O valor total pago nas emendas não é influenciado significativamente pela localidade do gasto.


```python
from statsmodels.formula.api import ols
model = ols('valor_empenhado ~ localidade_gasto', data=df_filtrado).fit()
```

```python
from statsmodels.stats.anova import anova_lm
anova_table = anova_lm(model, type=2)
print(anova_table)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

posthoc = pairwise_tukeyhsd(endog=df_filtrado['valor_empenhado'], groups=df_filtrado['localidade_gasto'], alpha=0.05)

print(posthoc)
```

```python
model = ols('valor_empenhado ~ regiao', data=df_filtrado).fit()
anova_table = anova_lm(model, type=2)
print(anova_table)
```
